```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ['movie id','movie title', 'release date','video release date',
              'IMDb URL' , 'unknown','Action',
'Adventure',
'Animation',
'Childrens',
'Comedy',
'Crime',
'Documentary',
'Drama',
'Fantasy',
'Film-Noir',
'Horror',
'Musical',
'Mystery',
'Romance',
'Sci-Fi',
'Thriller',
'War',
'Western']
item=pd.read_csv("u.item",encoding = "latin-1", sep = "|" , names =column_names)
item
```

```python
# create index : movie dictionary

index_movie_dict={}
for i in range(len(item)):
    index_movie_dict[i]=item.loc[i]['movie title']
index_movie_dict
```

```python
# create movie index dictionary

movie_index_dict={}
for i in range(len(item)):
    movie_index_dict[item.loc[i]['movie title']] =i
movie_index_dict
```

```python
# Drop unnecssary columns

item.drop(['movie id','movie title', 'release date','video release date',
              'IMDb URL'], axis = 1,inplace = True)
item
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(item)
similarity_matrix
```

```python
#  Generate recommendations 

recently_watched_movie = input("enter recently watched movie : ")
no_of_recommendations = int(input("enter no. of recommendations : "))
```

```python
idx = movie_index_dict[recently_watched_movie]
lst = []
for i in range(len(similarity_matrix)):
    lst.append((similarity_matrix[idx][i],i))
lst
```

```python
lst.sort(reverse = True)
lst
```

```python
recommendations = []
for i in range(len(lst)):
    recommendations.append(index_movie_dict[lst[i][1]])
recommendations
```

```python
recommendations.remove(recently_watched_movie)
recommendations[:no_of_recommendations]
```
